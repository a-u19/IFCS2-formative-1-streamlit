In this section, we are introducing Pandas by attempting to use its [documentation](https://pandas.pydata.org/docs/user_guide/10min.html#importing-and-exporting-data). 

```
import numpy as np
import pandas as pd
```

```
df2 = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)
```
