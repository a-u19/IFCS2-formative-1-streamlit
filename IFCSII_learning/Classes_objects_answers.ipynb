Create a class named Movie with attributes to hold information about movies.

Requirements:

- Define a class named Movie.
- The class should have two class attributes: title and director, with default values "Inception" and "Christopher Nolan" respectively.
- Define a method named play inside the class Movie, which prints "Now playing {title} by {director}!"
- After defining the class, create an instance of the class named my_movie.
- Print the values of the title and director attributes of the my_movie instance.
- Call the play method on the my_movie instance.

```
class Movie:  # Define a class named Movie
    # Define class attributes title and director with default values
    title = "Inception"
    director = "Christopher Nolan"
    
    # Define a method named play for the class Movie
    def play(self):
        # Print a message using the title and director attributes
        print(f"Now playing {self.title} by {self.director}!")

# Create an instance of the Movie class
my_movie = Movie()

# Print the title and director attributes of the my_movie instance
print(my_movie.title)  # This will print: Inception
print(my_movie.director)  # This will print: Christopher Nolan

# Call the play method on the instance of the Movie class
# This will print: Now playing Inception by Christopher Nolan!
my_movie.play()
```

### üèãÔ∏è  Practice Activity 3

Create a class named Student with a constructor and a method.

Requirements:

- Define a class named Student.
- The class should have a constructor that initializes two attributes: name and grade.
- Define a method named study inside the class Student, which prints "{Name} is studying hard!..."
- After defining the class, create an instance of the class named my_student.
- Call the study method on the my_student instance.

```
class Student:  # Define a class named Student
    def __init__(self, name, grade):  # Define the constructor with parameters name and grade
        self.name = name  # Initialize the name attribute with the provided name
        self.grade = grade  # Initialize the grade attribute with the provided grade
    
    def study(self):  # Define a method named study for the class Student
        print(f"{self.name} is studying hard, they have {self.grade}!")  # Print a message using the name attribute

# Create an instance of the Student class and initialize the attributes
my_student = Student("John Doe", "75")  

# Call the study method on the instance of the Student class
my_student.study()  # This will print: John Doe is studying hard!
```

### üèãÔ∏è  Practice Activity 4

Create a class named Rectangle with a constructor, setters, getters, and methods to calculate the area and perimeter.

Requirements:

- Define a class named Rectangle.
- The class should have a constructor that initializes two attributes: length and width.
- Define a getter and a setter for both the length and width attributes.
- Define a method named area inside the class Rectangle, which returns the area of the rectangle.
- Define another method named perimeter inside the class Rectangle, which returns the perimeter.
- After defining the class, create an instance of the class named my_rectangle, initializing the length and width attributes with any positive values.
- Use the getter methods to print the values of length and width, and call the area and perimeter methods to print the area and perimeter of the rectangle respectively.

```
class Rectangle:  # Define a class named Rectangle
    def __init__(self, length, width):  # Define the constructor with parameters length and width
        self.length = length  # Initialize the length attribute with the provided length
        self.width = width  # Initialize the width attribute with the provided width
    
    # Getter for the length attribute
    def get_length(self):
        return self.length
    
    # Setter for the length attribute
    def set_length(self, value):
        if value < 0:  # Perform validation
            raise ValueError("Length cannot be negative")
        self.length = value
    
    # Getter for the width attribute
    def get_width(self):
        return self.width
    
    # Setter for the width attribute
    def set_width(self, value):
        if value < 0:  # Perform validation
            raise ValueError("Width cannot be negative")
        self.width = value
        
    # Method to calculate the area of the rectangle
    def area(self):
        return self.length * self.width  # Calculate and return the area
    
    # Method to calculate the perimeter of the rectangle
    def perimeter(self):
        return 2 * (self.length + self.width)  # Calculate and return the perimeter


# Create an instance of the Rectangle class
my_rectangle = Rectangle(10, 5)

# Print the length and width attributes using the getter methods
print(f"Length: {my_rectangle.get_length()}")  # This will print: Length: 10
print(f"Width: {my_rectangle.get_width()}")  # This will print: Width: 5

# Call the area and perimeter methods and print the results
print(f"Area: {my_rectangle.area()}")  # This will print: Area: 50
print(f"Perimeter: {my_rectangle.perimeter()}")  # This will print: Perimeter: 30
```
